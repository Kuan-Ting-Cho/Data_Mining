```python
import nltk
import pandas as pd
import warnings
import numpy as np
pd.set_option("display.max_rows",10000)
warnings.filterwarnings('ignore')
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python
df = pd.read_csv('clean.csv').drop(columns=["s", "Unnamed: 6","total no.: 7987"])
df=df.dropna().reset_index(drop=True)
df.head(500)
```

```python
id_list=pd.unique(df['id'])
print(id_list)
```

```python
def whatever(dataframe,id,actual,label):
    part=dataframe[dataframe['id']==id]
    act_index=part[actual].index
    act_sents = nltk.sent_tokenize(part[actual][act_index[0]])
    
    lab_index=part[label].index
    overlap=[]
    for index in lab_index:
        lab_sents= nltk.sent_tokenize(part[label][index])
        start=0
        for j in range(len(lab_sents)):
            for k in range(start,len(act_sents)):
                if lab_sents[j] in act_sents[k]:
                    overlap.append(k)
                    start=k+1
                    break
    overlap=list(set(overlap))
    overlap.sort()
    return len(act_sents), overlap
```

```python
column=['id','q_length',"q'_overlap",'r_length',"r'_overlap"]
analysis = pd.DataFrame(columns=column)
for i in range(len(id_list)):
    len_q,lap_q=whatever(df,id_list[i],"q","q'")
    len_r,lap_r=whatever(df,id_list[i],"r","r'")
    id_df = pd.DataFrame([[id_list[i],len_q,lap_q,len_r,lap_r]], columns=column)
    analysis=analysis.append(id_df)
analysis=analysis.reset_index(drop=True)
analysis
```

```python
def whatever2(dataframe,id,actual,label):
    part=dataframe[dataframe['id']==id]
    act_index=part[actual].index
    act_sents = nltk.sent_tokenize(part[actual][act_index[0]])

    lab_index=part[label].index
    overlap=[]
    for index in lab_index:
        lab_sents= nltk.sent_tokenize(part[label][index])
        overlap.append(len(lab_sents))

    return len(act_sents), overlap
```

```python
column=['id','q_length',"q'_length",'r_length',"r'_length"]
analysis_length = pd.DataFrame(columns=column)
for i in range(len(id_list)):
    len_q,lap_q=whatever2(df,id_list[i],"q","q'")
    len_r,lap_r=whatever2(df,id_list[i],"r","r'")
    id_df = pd.DataFrame([[id_list[i],len_q,lap_q,len_r,lap_r]], columns=column)
    analysis_length=analysis_length.append(id_df)
analysis_length=analysis_length.reset_index(drop=True)
analysis_length
```

```python
test = pd.read_csv('test_clean.csv').drop(columns=["s"])
test.head(10)
```

```python
column=['id','q','r']
predict = pd.DataFrame(columns=column)
for i in range(len(test['id'])):
  q_sent = nltk.sent_tokenize(test['q'][i])
  r_sent = nltk.sent_tokenize(test['r'][i])

  if 7>=len(q_sent)>=2:
    a=q_sent[0]+' '+q_sent[1]
  elif len(q_sent)>=14:
    a=q_sent[0]+' '+q_sent[1]
  else:
    a=q_sent[0]

  if len(r_sent)==8 or len(r_sent)==11:
    b=r_sent[0]
  elif 14>=len(r_sent)>=2:
    b=r_sent[0]+' '+r_sent[1]
  elif len(r_sent)>=19:
    b=r_sent[0]+' '+r_sent[1]
  else:
    b=r_sent[0]

  pre = pd.DataFrame([[test['id'][i],a,b]], columns=column)
  predict=predict.append(pre)
```
